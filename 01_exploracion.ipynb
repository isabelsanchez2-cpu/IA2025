```python
z1 = pd.read_csv("train.csv")
z1
```

```python
print(z1.info())
```

```python
# Mostrar las primeras filas del DataFrame
print("Primeras 5 filas del DataFrame:")
display(z1.head())

# Mostrar información general del DataFrame (tipos de datos y valores no nulos)
print("\nInformación del DataFrame:")
z1.info()

# Mostrar estadísticas descriptivas de las columnas numéricas
print("\nEstadísticas descriptivas de columnas numéricas:")
display(z1.describe())

# Mostrar estadísticas descriptivas de las columnas de objetos (categóricas)
print("\nEstadísticas descriptivas de columnas categóricas:")
display(z1.describe(include='object'))

# Verificar valores faltantes por columna
print("\nValores faltantes por columna:")
display(z1.isnull().sum())
```

```python
z = z1[['F_EDUCACIONMADRE', 'E_VALORMATRICULAUNIVERSIDAD', 'RENDIMIENTO_GLOBAL']]
display(z.head())
```

```python
# Gráfico de barras para E_VALORMATRICULAUNIVERSIDAD
plt.figure(figsize=(12, 6))
sns.countplot(data=z, y='E_VALORMATRICULAUNIVERSIDAD', order=z['E_VALORMATRICULAUNIVERSIDAD'].value_counts().index)
plt.title('Conteo de E_VALORMATRICULAUNIVERSIDAD')
plt.xlabel('Conteo')
plt.ylabel('E_VALORMATRICULAUNIVERSIDAD')
plt.show()

# Gráfico de barras para F_EDUCACIONMADRE
plt.figure(figsize=(12, 8))
sns.countplot(data=z, y='F_EDUCACIONMADRE', order=z['F_EDUCACIONMADRE'].value_counts().index)
plt.title('Conteo de F_EDUCACIONMADRE')
plt.xlabel('Conteo')
plt.ylabel('F_EDUCACIONMADRE')
plt.show()
```
